```python
from datasets import load_dataset

dataset = load_dataset("tweets_hate_speech_detection")
```

```python
dataset['train'][0]
```

```python
from transformers import pipeline, set_seed
generator = pipeline('text-generation', model='gpt2')
set_seed(42)
# generator("Hello, I'm a language model,", max_length=30, num_return_sequences=5)
```

# Few Shot Experiment

1. Dari dataset toxic tadi, ambil 4 row
2. Susun setiap row jadi text.label dan concat semua row
3. Buang label terakhir
4. Input masukin generator
5. Check apa yang digenerate

```python
n_shot = 2
```

```python
input_text = ' [tweet:]'+dataset['train'][0]['tweet'] +' [label:]' +str(dataset['train'][0]['label'])

for i in range(n_shot):
    input_text += ' [tweet:]'+dataset['train'][i+1]['tweet'] +' [label:]' +str(dataset['train'][i+1]['label'])

# input_text += ' [tweet:]'+dataset['train'][13]['tweet'] +' [label:]'
```

```python
for i in range(10,20):
    if (dataset['train'][i]['label'] == 1):
        input_text += ' [tweet:]'+dataset['train'][i]['tweet'] +' [label:]' +str(dataset['train'][i]['label'])
input_text += ' [tweet:]'+dataset['train'][20]['tweet'] +' [label:]' +str(dataset['train'][20]['label'])
```

```python
input_text = input_text.replace('#', '')
input_text = input_text.replace('@user', '')
input_text
```

```python
input_text = input_text[:-1]
input_text
```

```python
generation_result = generator(input_text, max_length=100, num_return_sequences=5)
generation_result[0]
```
